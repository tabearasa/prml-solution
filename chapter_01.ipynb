### 1.1
We have to find a way to show that the given equations "express the same thing". The model $ y(x,\boldsymbol{w}) $ is given by the polynomial (1.1). Show that the coefficients $\boldsymbol{w} = \{w_{i}\}$ that minimize this error function are given by the solution to the following: $$ \sum_{j=0}^M A_{i,j}w_{j} = T_{i}$$ where $$ A_{i,j} = \sum_{n=1}^N (x_{n})^{(i+j)}  $$ $$T_{i} = \sum_{n=1}^N (x_{n})^it_{n} $$ 
Here a suffix i or j denotes the indes of a component, whereas $(x_{i})^i $ denotes $x$ raisesd to the power of $i$.

### Solution

```python
#Modeling of the "model"
grid = 1000
x_m = np.linspace(0,2,grid)
y_m = np.sin(2*np.pi*x_m)

def ground_truth(x):
    return np.sin(2*np.pi*x)

#Modeling of the data with noise
n_data = 10
noise_sd = 0.2
x_train = np.linspace(0,2,n_data)
y_train = np.sin(2*np.pi*x_train) + np.random.normal(loc=0, scale=noise_sd, size=n_data)
print("x_train: ", x_train, "\n y_train:", y_train)

plt.plot(x_m, y_m, 'g')
plt.scatter(x_train, y_train, s=80, facecolors='none', edgecolor='blue')
```

Formula (1.1): $ y(x,\boldsymbol{w}) = w_{0} + w_{1}x + w_{0}x² + ... + w_{M}x^M = \sum_{j=0}^M w_{j}x^j$

Formula (1.2): $ E(\boldsymbol{w}) = \frac{1}{2} \sum_{n=1}^N \{y(x_{n},w) - t_{n}\}²$

Formula (1.122): $ \sum_{j=0}^M A_{i,j}w_{j} = T_{i}$

```python
# Formula (1.1)
degree = 2
w = np.random.normal(loc=0, scale=noise_sd, size=degree) #initalizing some random weights
x = 2

def model(x,w):
    model = np.zeros(degree)
    for j in range(degree):
        model[j] = x**j
    model = w @ model
    return model

print("\n", "w = ", w, "\n x = ", x,"\n y(x,w) = ", model(x,w) )

#Formula (1.2) 
def error_function(y_train,x_train,w,n_data):
    error = np.zeros(n_data)
    for n in range(n_data):
        error[n] = (model(x_train[n],w) - ground_truth(y_train[n]))**2
        sum_error = 1/2 * np.sum(error)
    return sum_error

print("Error = ", error_function(y_train, x_train,w,n_data))

#Formula (1.122)
def A_ij(x_train,n_data,degree):
    A = np.zeros((n_data,degree))
    for n in range(n_data):
        for d in range(degree):
            A[n,d] = x_train[n]**(n+d)
    return A

#TODO
def T_i():
    return

print("Matrix A_ij: \n", A_ij(x_train, n_data, degree))
```

Now the Formula (1.122) is a different formulation of the goal to get the best fit trough a model with degree = M by minimizing this error function given in Formula (1.2). Therefore the Formula (1.2) have to be derived, with respect to w and set to zero. Take a look at the derivation:

```python
def plot_fit(degree):
    fig = plt.figure(figsize=(10,10*degree))

    for j in range(1,degree+1):
        ax = fig.add_subplot(degree*2.3,3,j)
        ax.set_title(f'Model fit with $M={j}$')
        plt.plot(x_m, y_m, 'g')
        plt.scatter(x_train, y_train, s=80, facecolors='none', edgecolor='blue')
        plt.text(1.5, 1, f'$M={j}$')
        plt.xlim([-0.1, 2.1])
        plt.ylim([-1.5, 1.5])
        
plot_fit(6)
```
